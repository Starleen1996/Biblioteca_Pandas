# DEMANDA DE ML

## ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos['Valor'].mean() # Aqui se pormedian todos los valores sin diversificar el tipo de propiedad
```

```python
# Agrupar los datos por tipos de datos para calcular el promedio de cada propiedad
#datos.groupby('Tipo').mean()

# Este parámetro numeric_only=True (disponible desde pandas 1.5) asegura que solo se usen columnas numéricas para el promedio.

datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# df_tipo_precio.plot(...) Utiliza el método .plot() de pandas, que se basa en matplotlib para crear gráficos a partir de un DataFrame o Series.
#kind='barh' Especifica el tipo de gráfico: 'barh' = gráfico de barras horizontales.
# figsize=(12,8) Define el tamaño del gráfico en pulgadas: 12 de ancho × 8 de alto.
# color='green' Asigna el color verde a las barras del gráfico.

df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind= 'barh',figsize=(12,8),color='green')
```

```python
datos.Tipo.unique() # Muestra los tipos de datos de la columna 'Tipo'
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salon', 'Casa Comercial',' Terreno Estandar',
                         'Cochera/Estacionamiento','Galpón/Depósio/Almacén','Tienda en Centro Comercial','Hotel','Lote/Condominio','Industria']
```

```python
datos.query('@inmuebles_comerciales in Tipo')
```

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df= datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind= 'barh',figsize=(12,8),color='green')
```

## ¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?"

```python
df.Tipo.unique()
```

```python
# Es demasiado util en el análisis de datos, permite contar los valores por tipo de inmueble
df.Tipo.value_counts()
```

```python
df.Tipo.value_counts(normalize=True) # Permite mostrar el porcentaje de cada tipo de inmueble
```
